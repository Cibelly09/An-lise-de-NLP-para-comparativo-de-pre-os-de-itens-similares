```python
import os
import pandas as pd
import glob
from fuzzywuzzy import process
from fuzzywuzzy import fuzz
from datetime import datetime, timedelta 
import re
```

```python
# üü© Defina aqui o per√≠odo desejado:
PERIODO_DIAS = 35  # Altere para 1, 7, 30, etc.
data_maxima = pd.to_datetime('today').normalize()
data_minima = data_maxima - pd.Timedelta(days=PERIODO_DIAS)
```

```python
# Fun√ß√£o para extrair a data do nome do arquivo (formato: YYYY_MM_DD)
def extrair_data_arquivo(nome_arquivo):
    match = re.search(r'(\d{4}_\d{2}_\d{2})', nome_arquivo)
    if match:
        return datetime.strptime(match.group(1), '%Y_%m_%d')
    return None
```

```python
# Caminho da pasta onde est√£o os arquivos
pasta_infomarket = r'K:\Pricing\Indicadores de Promo√ß√£o\Dados\Infomarket\Base detalhada'
pasta_smarket = r'K:\Pricing\Indicadores de Promo√ß√£o\Dados\Infomarket\Encartes Smarket'
```

```python
# Lista os arquivos que est√£o dentro do intervalo de tempo desejado
arquivos_validos = []
for nome_arquivo in os.listdir(pasta_infomarket):
    if nome_arquivo.endswith('.csv'):
        data_arquivo = extrair_data_arquivo(nome_arquivo)
        if data_arquivo and data_minima <= data_arquivo <= data_maxima:
            arquivos_validos.append(os.path.join(pasta_infomarket, nome_arquivo))

# Mostra os arquivos encontrados
print(f"üìÅ Arquivos encontrados ({len(arquivos_validos)}):")
for arquivo in arquivos_validos:
    print(" -", os.path.basename(arquivo))
```

```python
# Concatena os arquivos em um √∫nico DataFrame
df_consolidado = pd.concat([
    pd.read_csv(arquivo, encoding='utf-8', sep=',', low_memory=False)
    for arquivo in arquivos_validos
], ignore_index=True)
```

```python
# Salva o resultado em um novo CSV
output_path = r'K:\Pricing\Indicadores de Promo√ß√£o\Dados\Infomarket\Base2infomarket\baseinfo2.csv'
df_consolidado.to_csv(output_path, index=False, sep =';' , encoding='utf-8')

print(f"‚úÖ Arquivos consolidados salvos em: {output_path}")
print(f"üì¶ Total de registros combinados: {df_consolidado.shape[0]}")
```

```python
# Fun√ß√£o para obter o arquivo .csv mais recente de uma pasta
def obter_csv_mais_recente(pasta):
    arquivos_csv = [f for f in os.listdir(pasta) if f.endswith('.csv')]
    if not arquivos_csv:
        return None
    caminhos_completos = [os.path.join(pasta, f) for f in arquivos_csv]
    arquivo_mais_recente = max(caminhos_completos, key=os.path.getmtime)
    return arquivo_mais_recente

# Lendo os arquivos mais recentes de cada pasta
dfs = {}

for nome_origem, caminho_pasta in pastas.items():
    arquivo = obter_csv_mais_recente(caminho_pasta)
    if arquivo:
        print(f"Lendo arquivo mais recente da origem '{nome_origem}': {os.path.basename(arquivo)}")
        dfs[nome_origem] = pd.read_csv(arquivo, encoding='utf-8', sep=';')
    else:
        print(f"Nenhum arquivo CSV encontrado na pasta: {caminho_pasta}")
```

```python
from rapidfuzz import process, fuzz
import pandas as pd
from datetime import datetime
import os 
from tqdm import tqdm
```

```python
def encontrar_preco_concorrente(produto_smarket2, df_infomarket2, limite=60, n=2):
    """Busca os produtos mais parecidos na Infomarket pela descri√ß√£o e retorna at√© n resultados com pre√ßo e similaridade."""
    
    melhores = process.extract(
        produto_smarket2,
        df_infomarket2['produto'],
        scorer=fuzz.token_sort_ratio,
        limit=30  # traz os n melhores resultados
    )

    resultados = []
    
    for match_desc, score, _ in melhores:  # fuzz.token_sort_ratio tamb√©m compara palavras embaralhadas
        if score >= limite:
            linha = df_infomarket2[df_infomarket2['produto'] == match_desc].iloc[0]
            
            dados_concorrente = {
                'produto_concorrente': match_desc,
                'preco_concorrente': linha['price'],
                'data_inicio': linha['validity_start_date'],
                'data_fim': linha['validity_finish_date'],
                'nome_concorrente': linha['network'],
                'nome_encarte': linha['leaflet_name'],
                'link_encarte': linha['images_url'],
                'score_similaridade': score
            }
            # Verifica se j√° existe um concorrente com o mesmo encarte
            repetido = any(
                (r['nome_concorrente'] == dados_concorrente['nome_concorrente']) and
                (r['nome_encarte'] == dados_concorrente['nome_encarte'])
                for r in resultados
            )

            if not repetido: 
                resultados.append(dados_concorrente)
                
            if len(resultados) >= n:
                break # j√° temos os n √∫nicos desejados
    return resultados
    

# ---------- Execu√ß√£o principal ----------
if __name__ == "__main__":
    limite_similaridade = 60  # define o grau de similaridade m√≠nimo

    print("üîÑ Lendo base do Smarket...")
    df_smarket2.columns = df_smarket2.columns.str.strip().str.lower().str.replace(" ", "_")

    print("üîÑ Lendo bases da Infomarket...")
    df_infomarket2.columns = df_infomarket2.columns.str.strip().str.lower().str.replace(" ", "_")

    print("üîç Comparando produtos...")
    resultados = []

    for _, row in tqdm(df_smarket2.iterrows(), total=len(df_smarket2)):
        desc = row['produto']
        preco_smarket = row['meu_preco']
        concorrentes = encontrar_preco_concorrente(desc, df_infomarket2, limite_similaridade, n=2)
        print(f"\nüîç Produto Smarket: {desc}")
        print(f"   ‚Üí Concorrentes encontrados: {len(concorrentes)}")


        if concorrentes:
            for concorrente in concorrentes:
                resultados.append({
                    'produto': desc,
                    'preco_smarket': preco_smarket,
                    'produto_concorrente': concorrente['produto_concorrente'],
                    'preco_concorrente': concorrente['preco_concorrente'],
                    'nome_concorrente': concorrente['nome_concorrente'],
                    'nome_encarte': concorrente['nome_encarte'],
                    'link_encarte': concorrente['link_encarte'],
                    'data_inicio': concorrente['data_inicio'],
                    'data_fim': concorrente['data_fim'],
                    'score_similaridade': concorrente['score_similaridade'],
                    'diferenca_preco': preco_smarket - concorrente['preco_concorrente']
                })
        else:
            # se nenhum concorrente for parecido
            resultados.append({
                'produto': desc,
                'preco_smarket': preco_smarket,
                'produto_concorrente': None,
                'preco_concorrente': None,
                'nome_concorrente': None,
                'nome_encarte': None,
                'link_encarte': None,
                'data_inicio': None,
                'data_fim': None,
                'score_similaridade': None,
                'diferenca_preco': None
            })

    df_resultado = pd.DataFrame(resultados)

    # caminho para salvar o arquivo
    pasta_destino = r"K:\Pricing\Indicadores de Promo√ß√£o\Dados\Infomarket\Automatizacao\COMPARATIVO"
    os.makedirs(pasta_destino, exist_ok=True)

    # Exporta o resultado
    nome_arquivo = f"comparativo_precos_{datetime.today().strftime('%Y%m%d')}.xlsx"
    caminho_completo = os.path.join(pasta_destino, nome_arquivo)

    df_resultado.to_excel(caminho_completo, index=False)

    print(f"\n‚úÖ Compara√ß√£o conclu√≠da! Resultado salvo em:\n{caminho_completo}")
```
